```python
import sqlite3
import pandas as pd

# Create a NEW database
conn = sqlite3.connect("FoodDelivery_New.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE FoodDelivery (
    OrderID INTEGER PRIMARY KEY,
    CustomerName TEXT,
    RestaurantName TEXT,
    FoodItem TEXT,
    Quantity INTEGER,
    Price REAL,
    PaymentMethod TEXT,
    DeliveryPartner TEXT,
    OrderStatus TEXT,
    OrderDate TEXT
)
""")

conn.commit()
print("Table created successfully.")
```
