Python Code for Ecommersce Transaction Dataset

```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Number of rows
num_rows = 10

# Create dataset with exactly 10 columns
data = {
    'TransactionID': np.arange(1001, 1001 + num_rows),
    'CustomerID': np.random.randint(101, 150, num_rows),
    'ProductID': np.random.randint(2001, 2050, num_rows),
    'Quantity': np.random.randint(1, 5, num_rows),
    'Price': np.round(np.random.uniform(10.0, 500.0, num_rows), 2),
    'TransactionDate': pd.date_range(start='2023-01-01', periods=num_rows),
    'PaymentMethod': np.random.choice(['Credit Card', 'Debit Card', 'UPI', 'PayPal'], num_rows),
    'ShippingAddress': [f'Address {i}' for i in range(1, num_rows + 1)],
    'OrderStatus': np.random.choice(['Delivered', 'Pending', 'Cancelled'], num_rows),
    'Category': np.random.choice(['Electronics', 'Books', 'Clothing', 'Home Appliances'], num_rows)
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the dataset
display(df)

# Print dataset shape
print("Dataset Shape:", df.shape)
```
